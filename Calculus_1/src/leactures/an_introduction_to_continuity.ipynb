## Formulas and Properties of Continuity:

f is continues if.

```julia
f() ~ f() + g() ~ f() - g() ~ f()*g() ~ f() // g(x != 0)
```

**This means also that the composition of a continuous function is another continuous function, so closer under continuous function.**

## Example 1: A Removable Discontinuity. 

```julia
fx(x) = x^2 - 4 // x - 2
```

```julia
ax = restart_axis()
```

```julia
lines( [1:2;], map(x -> fx(x), 1:2))
scatter!( [1:2;], map(x -> fx(x), 1:2))
current_figure()
```

```julia
@syms x::Real
fx(x) = sin(x) / x
```

Discontinue as zero is undefine.

```julia
limit(substitute(fx(x), (x => 0)), x, 0)
```

```julia
map(v -> limit(substitute(fx(x), (x => v)), x, 0), -5:5)
```

```julia
lines(map(v -> limit(substitute(fx(x), (x => v)), x, 0), -100:100))
```

```julia
lines( [0:25;], map(v -> limit(substitute(fx(v), (x => v)), x, 0), 0:25))
scatter!( [-25:0;], map(x -> fx(x), -25:0))
current_figure()
```

## Example 2: An Application of the Intermediate Value Theorem.

```julia
fx(x) = x^3 + 2x - 1 
 f(x) ~ fx(x)
```

```julia
expand(fx(x))
```

```julia
substitute(fx(x), (x => 0))
```

```julia
limit(substitute(fx(x), (x => 0)), x, 1)
```

```julia
map(v -> limit(substitute(fx(v), (x => v)), x, 1), -5:5) 
```

```julia
scatterlines(map(v -> limit(substitute(fx(v), (x => v)), x, 1), -5:5))
```

```julia
f(c) ~ 0 ~ c ~ limit(substitute(fx(x), (x => 6)), x, 1) * 2 / 1000
```

```julia
fx(x) = 6 / x
```

```julia
scatterlines(map(v -> limit(substitute(fx(x), (x => v)), x, 0), -5:5))
```
